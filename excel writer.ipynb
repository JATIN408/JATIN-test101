```python
# import required module 
import xlsxwriter 
import pandas as pd
import math


#File Name
excel_file = 'test_excel.xlsx'   

#Variable Used 
s_no = 1
temp_list = []
final_list = []
scenarios_temp = []
file_name = excel_file.split('_')
#project_name = file_name[0]
#project_date = (file_name[2].split('.'))[0]
first_run = True

#Read data from file 
i_excel_file = pd.read_excel(excel_file,keep_default_na=False)
x = i_excel_file.to_dict(orient='record')


# Total no of record 
t = i_excel_file.shape # Tuple hold shape of excel 
no_of_record = t[0]


#Function to convert list to string 
def listToString(s):  
	str1 = " "  
	for ele in s:  
		str1 += ele   
	return str1


for i in x:
	p_s_no = i['Sl No']
	p_scenarios = i['Scenarios']
	p_status = i['Pass/Fail']
	x = i['Comment']
	
	
	if str(p_scenarios)=='':
		p_scenarios = 'No Test Case'
		test_id = 'No Test Case'
	else:
		test_id_split = p_scenarios.split('_')    #Scenarios -- Test ID  -- Split with '_'
		
		#test_id = ((test_id_split[0]).split(']'))[1]
		t_scenario = str([' '.join(test_id_split[1:])])
		scenario = listToString(test_id_split[1:])
		
	if str(p_status)=='Failed':
		p_status = 'Fail'

	if str(p_status)=='':
		p_status = 'No Status'	
	
	if str(x)=='':
		total_no_of_comment = ['No Comment']
		
	else:
		total_no_of_comment = x.split('\n')    #Total comment in list 
	no_of_comment = len(total_no_of_comment)
		
	
	if no_of_comment>=1:
		for j in total_no_of_comment:
			temp_list.append(s_no)
			temp_list.append(test_id)
			temp_list.append(scenario)
			temp_list.append(p_status)
			temp_list.append(j)
			s_no=s_no+1
			x = temp_list.copy()
			final_list.append(x)
			temp_list.clear()


#Final Data Stored in Tuple 
final_tuple = tuple(final_list)


#Workbook to do data write in excel
#Project Name 
workbook = xlsxwriter.Workbook('New_test_Case'+'.xlsx') 
  
# By default worksheet names in the spreadsheet will be  
# Sheet1, Sheet2 etc., but we can also specify a name. 
worksheet = workbook.add_worksheet("TC") 
    
# Start from the first cell. Rows and 
# columns are zero indexed. 
row = 0
col = 0
  
  
if first_run:
		worksheet.write(row, col, 'Project') 
		worksheet.write(row, col + 1, 'Release') 
		worksheet.write(row, col + 2, 'TC')
		worksheet.write(row, col + 3, 'TC Description')
		worksheet.write(row, col + 4, 'Status')
		worksheet.write(row, col + 5, 'Reason')
		row += 1
		first_run = False  
  
# Iterate over the data and write it out row by row. 
for id,test_name,scenarios, status,comment in (final_tuple): 
	worksheet.write(row, col, 'Test') 
	worksheet.write(row, col + 1, 'Today') 
	worksheet.write(row, col + 2, test_name)
	worksheet.write(row, col + 3, scenarios)
	worksheet.write(row, col + 4, status)
	worksheet.write(row, col + 5, comment)
	row += 1
  
workbook.close() 

#Final Message 
print('Succeed...')
```
